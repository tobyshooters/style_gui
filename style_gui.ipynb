# Adaptive Style Transfer with AdaIN

```python
from ipywidgets import FloatSlider, Checkbox
from ipyfilechooser import FileChooser
from gui import *
import os
```

### Setup the Style Transfer

```python
# Choose content file
content_fc = FileChooser('./input/content', title="Choose Your Image")
display(content_fc)

# Choose style file
style_fc = FileChooser('./input/style', title="Choose Your Style")
display(style_fc)

# Choose alpha
alpha_slider = FloatSlider(value=0.5, min=0, max=1.0, step=0.05, description='Alpha:', indent=False)
print("Higher alpha means more style, less content")
display(alpha_slider)

# Preserve color
preserve_box = Checkbox(value=False, description='Preserve Color', indent=False)
display(preserve_box)
```

### Run the Neural Network

```python
command = "python3 src/test.py --content {} --style {} --alpha {}".format(
    content_fc.selected,
    style_fc.selected,
    alpha_slider.value
)

if preserve_box.value:
    command += " --preserve_color"

print("Running style transfer...")
print("> " + command)
os.system(command)
print("Done")
```
