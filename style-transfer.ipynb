### Adaptive Style Transfer with AdaIN

```python
from gui import *
from ipyfilechooser import FileChooser
import os
```

```python
content_fc = FileChooser('./input/content', title="Choose Your Image")
display(content_fc)

style_fc = FileChooser('./input/style', title="Choose Your Style")
display(style_fc)
```

```python
preserve_color = False
alpha = 0.5 # Between 0 and 1

command = "python src/test.py --content {} --style {} --alpha {}".format(
    content_fc.selected,
    style_fc.selected,
    alpha
)

if preserve_color:
    command += " --preserve_color"

print(command)
os.system(command)
```

```python
result_path = get_result_path(content_fc.selected, style_fc.selected)
display_image(result_path)
```
